```python
from pathlib import Path
import os
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

import numpy as np
import matplotlib

import PIL
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rcParams.update(
    {
    'text.usetex': True,
    "font.family": "serif",
    "pgf.texsystem": "pdflatex",
    "pgf.rcfonts" : False,
    }
)
```

```python
pt_to_in = 1/72.27

# this needs to be set to the desired plot width in points
# for example can be determined using "\showthe\textwidth" in latex
page_width = 345 

figure_width = page_width * pt_to_in
figure_width = np.round(figure_width, 2)
figure_height = 2.5
figure_width, figure_height
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
f, ax = plt.subplots(figsize=(figure_width, figure_height), dpi=150)
ax.plot(x, y1, label="y1")
ax.plot(x, y2, label="y2")

ax.legend(frameon=False)
ax.set_ylabel("x")
ax.set_ylim(-1, 1)
ax.set_xlabel("y axis with a latex math piece $N_{\mathrm{r}}$")
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
ax.xaxis.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

f.tight_layout()
f.savefig(f"demo-plot.pgf")
```
